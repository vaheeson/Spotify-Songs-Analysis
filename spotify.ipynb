```python
df = df.drop_duplicates()
```

```python
#genre analysis
top_genres = df['playlist_genre'].value_counts()

print(top_genres)
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df,
              y='playlist_genre',
              order=df['playlist_genre'].value_counts().index)

plt.title("Number of Songs by Genre")
plt.show()
```

```python
#Artist analysis
top_artists = df['track_artist'].value_counts().head(10)

print(top_artists)
```

```python
top_artists.plot(kind='bar')

plt.title("Top 10 Artists")
plt.ylabel("Number of Songs")
plt.show()
```

```python
#Popularity Analysis
plt.figure(figsize=(8,5))

sns.histplot(df['track_popularity'],
             bins=30)

plt.title("Track Popularity Distribution")
plt.show()
```

```python
#Average popularity by Genre
genre_popularity = df.groupby(
    'playlist_genre'
)['track_popularity'].mean()

print(genre_popularity)
```

```python
#Audio feature Analysis
audio_features = [
    'danceability',
    'energy',
    'speechiness',
    'acousticness',
    'instrumentalness',
    'liveness',
    'valence'
]
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(
    df[audio_features].corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Audio Features Correlation")
plt.show()
```

```python
#Danceability vs Popularity
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x='danceability',
    y='track_popularity'
)

plt.title("Danceability vs Popularity")
plt.show()
```

```python
#Energy By Genre
plt.figure(figsize=(10,6))

sns.boxplot(
    data=df,
    x='playlist_genre',
    y='energy'
)

plt.xticks(rotation=45)
plt.title("Energy Distribution by Genre")
plt.show()
```

```python
#Calculate average matrics
df.groupby('playlist_genre')[[
    'track_popularity',
    'danceability',
    'energy'
]].mean().sort_values(
    'track_popularity',
    ascending=False
)
```

```python
df.to_csv(
    "spotify_cleaned.csv",
    index=False
)
```
